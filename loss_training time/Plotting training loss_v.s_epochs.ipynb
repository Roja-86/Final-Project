```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_paths = {
    "Custom Model": "D:/USC_Course/CSCE 790 Section 007 Neural Networks and Their Applications/atex-main/loss_training time/training_metrics_custom_model.csv",
    "EfficientNet B0": "D:/USC_Course/CSCE 790 Section 007 Neural Networks and Their Applications/atex-main/loss_training time/training_metrics_efficientnet_b0.csv",
    "EfficientNet B7": "D:/USC_Course/CSCE 790 Section 007 Neural Networks and Their Applications/atex-main/loss_training time/training_metrics_efficientnet_b7.csv",
    "ShuffleNet V2": "D:/USC_Course/CSCE 790 Section 007 Neural Networks and Their Applications/atex-main/loss_training time/training_metrics_shufflenet_v2.csv"
}

# Font sizes
title_font_size = 16
label_font_size = 14
tick_font_size = 12
legend_font_size = 12

# Plot training loss for each model
plt.figure(figsize=(10, 6))

for model, path in file_paths.items():
    # Load and process each file
    df = pd.read_csv(path)
    df = df[['epoch', 'train_loss']].dropna()
    df['epoch'] = pd.to_numeric(df['epoch'], errors='coerce')
    df['train_loss'] = pd.to_numeric(df['train_loss'], errors='coerce')

    # Plot each model
    plt.plot(df['epoch'], df['train_loss'], label=model)

# Finalize the plot
plt.title('Training Loss Over Epochs for All Models', fontsize=title_font_size)
plt.xlabel('Epoch', fontsize=label_font_size)
plt.ylabel('Training Loss', fontsize=label_font_size)
plt.legend(fontsize=legend_font_size)
plt.grid()

# Customize tick font size
plt.xticks(fontsize=tick_font_size)
plt.yticks(fontsize=tick_font_size)

# Save the plot
output_path = "D:/USC_Course/CSCE 790 Section 007 Neural Networks and Their Applications/atex-main/loss_training time/training_loss_plot.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')
plt.show()
```
